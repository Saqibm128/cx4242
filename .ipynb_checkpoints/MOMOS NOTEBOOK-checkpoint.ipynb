```python
import multiprocessing as mp
```

```python
manager = mp.Manager()
to_run = manager.Queue()
```

```python
def run(in_q):
    for node_id in iter(in_q.get, None):
        try:
            node_data = api.NodeGet(node_id)
            print(node_data)
            read_sql("update OSM_nodes set latitude={}, longitude={} where osm_node_id={}".format(node_data["lat"], node_data["lon"], node_id))
        except Exception as e:
            print(e)
```

```python
in_q = manager.Queue()
[in_q.put(node_id) for node_id in all_osm_nodes["osm_node_id"]]
[in_q.put(None) for i in range(20)]
```

```python
p = [mp.Process(target=run, args=(in_q,)) for i in range(20)]
```

```python
[process.start() for process in p]
```
